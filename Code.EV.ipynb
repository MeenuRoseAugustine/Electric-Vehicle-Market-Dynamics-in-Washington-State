```python
#checking unique values per column
print("Unique values per column:", df.nunique())
```

```python
#checking missing values per column
print("Missing values per column:", df.isnull().sum())
```

```python
#fill small no.of missing categorical values with mode
for col in['County','City','Postal Code','Electric Utility','2020 Census Tract']:
    if df[col].isnull().sum()>0:
        df[col]=df[col].fillna(df[col].mode()[0])
```

```python
#fill missing values with median
for col in['Electric Range','Base MSRP']:
    if df[col].isnull().sum()>0:
        df[col]=df[col].fillna(df[col].median())
```

```python
#drop rows of missing values in Legislative District
df=df[df['Legislative District'].notna()]
```

```python
#drop rows with missing data in vehicle location
df=df.dropna()
```

```python
#removing duplicates
df=df.drop_duplicates()
```

```python
#checking current data types
print(df.dtypes)
```

```python
#change datatypes
#postalcode datatype float64 to string
df['Postal Code']=df['Postal Code'].astype('Int64').astype(str)
#Legislative Distric datatype float64 to int
df['Legislative District']=df['Legislative District'].astype('Int64')
```

```python
#Object datatypes are changing to category
cat_cols=['County','City','State','Make','Model','Electric Vehicle Type','Clean Alternative Fuel Vehicle (CAFV) Eligibility','Electric Utility']
for col in cat_cols:
    df[col]=df[col].astype('category')
```

```python
#clean & standardize City & County names
df['City']=df['City'].str.strip().str.title()
df['County']=df['County'].str.strip().str.title()
```

```python
#new columns
#Derived columns -is electric vehicle or not
df['Is_Electric']=df['Electric Vehicle Type'].notnull().astype(int)
```

```python
#checking vehicle age
df['Vehicle_Age']=2025-df['Model Year']
```

```python
#checking high range EVs
df['High_Range_EVs']=df['Electric Range']>200
```

```python
df['Price_per_Mile']=df.apply(lambda X:X['Base MSRP']/ X['Electric Range'] if X['Electric Range'] > 0
                              else None,
                              axis=1
                             )
```

```python
#US states into board regions
def get_region(state): 
    west = ['WA', 'OR', 'CA', 'NV'] 
    east = ['NY', 'NJ', 'MA', 'PA'] 
    south = ['TX', 'FL', 'GA'] 
    midwest = ['IL', 'OH', 'MI'] 
     
    if state in west: 
        return 'West'
    elif state in east: 
        return 'East' 
    elif state in south: 
        return 'South' 
    elif state in midwest: 
        return 'Midwest'
    else: 
        return 'Other' 
 
df['Region'] = df['State'].apply(get_region)
```

```python
df
```

```python
#Check for missing or non-numeric values to avoid NaN
print(df[['Base MSRP', 'Electric Range']].info())
print(df[['Base MSRP', 'Electric Range']].head())
```

```python
#finding rows with actual values
print((df['Base MSRP'] > 0).sum())
print((df['Electric Range'] > 0).sum())
print(((df['Base MSRP'] > 0) & (df['Electric Range'] > 0)).sum())
```

```python
df = df[(df['Base MSRP'] > 0) & (df['Electric Range'] > 0)].copy()
```

```python
#price per mile calculation
df['Price_per_Mile'] = df['Base MSRP'] / df['Electric Range']
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set(style="whitegrid")
```

```python
#univariate analysis
num_cols = ['Electric Range', 'Base MSRP', 'Price_per_Mile', 'Vehicle_Age'] 
df[num_cols].describe()
```

```python
#univariate analysis
for col in num_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=col, hue='Electric Vehicle Type', kde=True, bins=30, palette='Set2', alpha=0.5, multiple='layer')
    plt.title(f'Distribution of {col} by EV Type (BEV vs PHEV)')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.tight_layout()
    plt.show()
```

```python
#boxplots
for col in num_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x='Electric Vehicle Type', y=col)
    plt.title(f'Boxplot of {col} by Electric Vehicle Type')
    plt.xlabel('Electric Vehicle Type')
    plt.ylabel(col)
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.tight_layout()
    plt.show()
```

```python
#2.bivariate analysis
#correlation
for ev_type in df['Electric Vehicle Type'].unique():
    plt.figure(figsize=(8, 6))
    subset = df[df['Electric Vehicle Type'] == ev_type]
    corr_subset = subset[num_cols].corr()
    sns.heatmap(corr_subset, annot=True, cmap='coolwarm', center=0, linewidths=0.5, fmt=".2f")
    plt.title(f'Correlation Matrix for {ev_type}')
    plt.xticks(rotation=45)
    plt.yticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#Price_per_Mile vs. Electric Range 
#scatterplot
sns.scatterplot(data=df, x='Electric Range', y='Price_per_Mile', hue='High_Range_EVs') 
plt.title('Price per Mile vs Electric Range') 
plt.show()
```

```python
#Price per Mile by Make (mean)
df.groupby('Make')['Price_per_Mile'].mean().sort_values(ascending=False).head(10).plot(kind='barh') 
plt.title('Avg. Price per Mile by Make') 
plt.xlabel('Price per Mile') 
plt.show()
```

```python
#3.Multivariate Analysis
#Pairwise Plot
sns.pairplot(df[['Base MSRP', 'Electric Range', 'Price_per_Mile', 'Vehicle_Age']]) 
plt.suptitle('Pairplot of Key Features', y=1.02) 
plt.show()
```

```python
#Region v/s Price per Mile Boxplot
plt.figure(figsize=(12, 6)) 
sns.boxplot(x='Region', y='Price_per_Mile', data=df) 
plt.title('Price per Mile by Region') 
plt.xticks(rotation=45) 
plt.show()
```

```python
#Count of EVs by Make
df['Make'].value_counts().head(10).plot(kind='barh') 
plt.title('Top 10 EV Makes') 
plt.xlabel('Count') 
plt.show()
```

```python
#4.Groupby Region & CAFV Eligibility
cafv_region = df.groupby(['Region', 'Clean Alternative Fuel Vehicle (CAFV) Eligibility']).size().unstack() 
cafv_region.plot(kind='bar', stacked=True, figsize=(10, 6)) 
plt.title('CAFV Eligibility by Region') 
plt.ylabel('Number of Vehicles') 
plt.show()
```

```python
#5.Pivot Tables and Aggregation
#Average MSRP and Range by Model Year
pivot = pd.pivot_table(
    df,
    values=['Base MSRP', 'Electric Range'],
    index=['Model Year', 'Electric Vehicle Type'],
    aggfunc='mean',
    observed= True
)

print(pivot)
```

```python
#2.Base MSRP Distribution
plt.figure(figsize=(10, 5)) 
sns.histplot(bevs['Base MSRP'], color='blue', label='BEV', kde=True, bins=30) 
sns.histplot(phevs['Base MSRP'], color='orange', label='PHEV', kde=True, bins=30) 
plt.title('Base MSRP Distribution: BEV vs PHEV') 
plt.xlabel('Price ($)') 
plt.ylabel('Count') 
plt.legend() 
plt.show()
```

```python
#3.Scatter Plot – Price vs Range
sns.scatterplot(data=df, x='Electric Range', y='Base MSRP', hue='Electric Vehicle Type') 
plt.title('Base MSRP vs Electric Range by EV Type')
plt.xlabel('Electric Range (miles)') 
plt.ylabel('Base MSRP ($)') 
plt.show()
```

```python
#6.Subplot – Range and Price by EV Type
fig, axs = plt.subplots(1, 2, figsize=(14, 6)) 
sns.boxplot(data=df, x='Electric Vehicle Type', y='Electric Range', ax=axs[0]) 
axs[0].set_title('Electric Range by EV Type') 
sns.boxplot(data=df, x='Electric Vehicle Type', y='Base MSRP',
ax=axs[1]) 
axs[1].set_title('Base MSRP by EV Type') 
plt.tight_layout()
plt.show()
```
